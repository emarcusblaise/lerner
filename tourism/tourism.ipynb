```python
import pandas as pd
```

```python
filename = 'Monthly-Arrivals-2000-Present.xlsx'
countries_df = pd.read_excel(filename)
```

```python
countries_df = pd.read_excel(filename,
                             skiprows=20,
                             header=None,
                             usecols=[1,2],
                             names=['country', 'region'],
                             index_col=0).dropna()
```

```python
countries_df
```

```python
travel_df = pd.read_excel(filename).drop(range(0,19))
```

```python
travel_df = travel_df.drop([1, ' ', 'World \nRegion', 'Notes:'], axis='columns')
```

```python
travel_df.columns[0]
```

```python
travel_df = travel_df.rename(columns={travel_df.columns[0]: 'country'})
```

```python
travel_df 
```

```python
travel_df.index.isna()
```

```python
travel_df = travel_df.loc[~travel_df.index.isna()]
```

```python
travel_df.columns
```

```python
drop_cols = [one_name for one_name in travel_df.columns if str(one_name).startswith("Unnamed")]
```

```python
travel_df = travel_df.drop(drop_cols, axis='columns')
```

```python
rename_dict = {one_name: one_name.removesuffix('Preliminary').strip() for one_name in travel_df.columns if str(one_name).endswith('Preliminary')}
```

```python
travel_df = travel_df.rename(columns=rename_dict)
```

```python
travel_df.columns = pd.to_datetime(travel_df.columns)
```

```python
travel_df = travel_df.replace(' -   ', 0)
```

```python
travel_df = travel_df.fillna(0)
```

```python
travel_df = travel_df.astype(int)
```
