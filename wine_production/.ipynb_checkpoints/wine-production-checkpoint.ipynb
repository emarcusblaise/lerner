```python
import pandas as pd
```

```python
filename = 'data.xlsx'

df = (
    pd
    .read_excel(filename)
)
```

```python
df
```

```python
df = (
    pd
    .read_excel(filename)
    .filter(regex='^[\w/]+$', axis='columns')
)
```

```python
df = (
    pd
    .read_excel(filename)
    .filter(regex='^[\w/]+$', axis='columns')
    .iloc[:-2]
)
```

```python
df
```

```python
df['Unit'].value_counts()
```

```python
df.loc[
    df['Product'] == 'Wine', 
    'Unit'
].value_counts()
```

```python
df.loc[
    df['Product'] == 'Wine', 
    'Unit'
].value_counts(normalize=True)
```
